# Offline Evaluation

## Step 1: Simulate Ranking Pairs of E and P

```python
def create_ranking_pairs():
    """
    Outputs a list of tuples with all distinct ranking pairs using a pool of 12 document ids.
    The first list in each pair corresponds to system E, the second to system P.
    """
    pairs = []
    
    # add pairs without overlap
    for rel_list in list(product([0, 1], repeat=6)): # all orders of relevance
        doc = [-1, 5] # unused doc id with relevance 0 and 1
        l_P = []
        l_E = []
        for r in rel_list[:3]:
            doc[r] += 1
            l_P.append(doc[r])
        for r in rel_list[-3:]:
            doc[r] += 1
            l_E.append(doc[r])
        pairs.append((l_P, l_E))
    
    # add pairs with one overlap
    for overlap in list(product([0, 1, 2], repeat=2)): # all possible overlaps [pos in list P, pos in list E]
        for rel_list in list(product([0, 1], repeat=5)): # all orders of relevance (excluding overlapping docs)
            doc = [-1, 5] # unused doc id with relevance 0 and 1
            l_P = [-1,-1,-1]
            l_E = [-1,-1,-1]
            i = 0
            # construct list P, and overlapping doc of list E
            for i in range(3):
                r = rel_list[i]
                doc[r] += 1
                l_P[i] = doc[r]
                if i == overlap[0]:
                    l_E[overlap[1]] = doc[r]
            # construct remainder of list E
            j = 0
            for i in range(3):
                if l_E[i] == -1:
                    r = rel_list[3+j]
                    doc[r] += 1
                    l_E[i] = doc[r]
                    j += 1
            pairs.append((l_P, l_E))
                
    # add pairs with two overlaps
    overlaps = list(product([0, 1, 2], repeat=4)) # all possible double overlaps as before, concatenated
    overlaps = [o for o in overlaps if o[0]!=o[2] and o[1]!=o[3] and o[0]<o[2]] # remove impossible overlaps
    for overlap in overlaps:
        for rel_list in list(product([0, 1], repeat=4)): # all orders of relevance (excluding overlapping docs)
            doc = [-1, 5] # unused doc id with relevance 0 and 1
            l_P = [-1,-1,-1]
            l_E = [-1,-1,-1]
            # construct list P, and overlapping doc of list E
            for i in range(3):
                r = rel_list[i]
                doc[r] += 1
                l_P[i] = doc[r]
                if i == overlap[0]:
                    l_E[overlap[1]] = doc[r]
                if i == overlap[2]:
                    l_E[overlap[3]] = doc[r]
            # construct remainder of list E
            j = 0
            for i in range(3):
                if l_E[i] == -1:
                    r = rel_list[3+j]
                    doc[r] += 1
                    l_E[i] = doc[r]
                    j += 1
            pairs.append((l_P, l_E))

    # add pairs with three overlaps
    overlaps = list(product([0, 1, 2], repeat=6)) # all possible double overlaps as before, concatenated
    overlaps = [o for o in overlaps if o[0]!=o[2]!=o[4]!=o[0] and o[1]!=o[3]!=o[5]!=o[1] and  o[0]<o[2]<o[4]] # remove impossible overlaps
    for overlap in overlaps:
        for rel_list in list(product([0, 1], repeat=3)): # all orders of relevance (excluding overlapping docs)
            doc = [-1, 5] # unused doc id with relevance 0 and 1
            l_P = [-1,-1,-1]
            l_E = [-1,-1,-1]
            # construct list P, and overlapping doc of list E (so all of them)
            for i in range(3):
                r = rel_list[i]
                doc[r] += 1
                l_P[i] = doc[r]
                if i == overlap[0]:
                    l_E[overlap[1]] = doc[r]
                if i == overlap[2]:
                    l_E[overlap[3]] = doc[r]
                if i == overlap[4]:
                    l_E[overlap[5]] = doc[r]
            pairs.append((l_P, l_E))
        
    return pairs

def get_relevance(doc):
    """
    Outputs the integer relevance corresponding to a document id
    """
    if (doc<0 or doc>11):
        raise Exception('Document id should be from 0 to 12!')
    if doc<6:
        return 0
    else:
        return 1
    
def get_relevance_list(docs):
    """
    Maps a list of document ids to integer relevance
    """
    return [get_relevance(doc) for doc in docs]
```

Note that create_ranking_pairs() creates all distinct pairs, that is: all distinct combinations of documents shared between the two lists (including no shared documents) and relevance scores.

## Step 2: Compute the $\Delta$ERR and form bins

```python
def compute_click_probability(rel):
    """
    Outputs the click-probability theta.
    """
    rel_max = 1
    return (2**rel - 1)/2**rel_max

def compute_ERR(list):
    """
    Computes Expected Reciprocal Rank for a list of form [doc_id1, doc_id2, doc_id3].
    """
    ERR = 0
    for r in range(len(list)):
        theta_r = compute_click_probability(get_relevance(list[r]))
        prod = 1
        for i in range(r):
            prod *= (1-compute_click_probability(get_relevance(list[i])))
        ERR += prod*theta_r*1/(r+1)
    return ERR

def divide_pairs_over_bins(list_pairs):
    """
    Divides a list of pairs over 10 bins according to the delta ERR.
    The i'th element of dERRs contains all pairs with delta EER between i/10 and (i+1)/10
    Note that the first bin only starts at 0.05 to make sure it does not happen too often that P outperforms E
    """
    dERRs = [[] for i in range(10)]
    for i in list_pairs:
        deltaERR = compute_ERR(i[0]) - compute_ERR(i[1])
        #Keep all the positive delta ERRs and put them in the respective bin
        if deltaERR > 0.05:
            if deltaERR < 0.1:
                dERRs[0].append(i)
            else:
                bin = str(deltaERR*10)
                dERRs[int(bin[0])].append(i)
    return dERRs
```

# Online Evaluation
## Step 3: Implement Team-Draft and Probabilistic Interleaving

```python
def coin_to_ranker(cointoss):
    """
    Maps zero to E (experimental system) and one to P (production system)
    """
    if (cointoss == 0):
        return "E"
    if (cointoss == 1):
        return "P"
    else:
        raise Exception('This number should be either 0 or 1!')
        
def softmax(r, max_r):
    """
    The softmax takes the rank of a document as well as the maximal rank and returns
    the probability of choosing that document
    """
    tau = 3
    ranks = range(1,max_r+1)
    normalizer = sum([rank**(-tau) for rank in ranks])
    return r**(-tau)/normalizer

def td_interleave(pair):
    """
    Performs team draft interleaving to a ranking pair
    Returns interleaved list as [(doc_id1, S1), (doc_id2, S2), (doc_id3, S3)] where
    Si describes which list contributed the document: E or P
    """
    new_pair = deepcopy(list([pair[0], pair[1]]))
    result = []
    for i in range(3):
        cointoss = random.randint(0, 1)
        result += [(new_pair[cointoss][0], coin_to_ranker(cointoss))]
        if (new_pair[cointoss][0] in new_pair[1 - cointoss]):
            new_pair[1 - cointoss].remove(new_pair[cointoss][0])
        new_pair[cointoss] = new_pair[cointoss][1:]
    return result

def pr_interleave(pair):
    """
    Performs probabilistic interleaving to a ranking pair
    Returns interleaved list as [(doc_id1, S1), (doc_id2, S2), (doc_id3, S3)] where
    Si describes which list contributed the document: E or P
    """
    new_pair = deepcopy(list([pair[0], pair[1]]))
    result = []
    sm = [[],[]]
    sm[0] = [softmax(r, 3) for r in range(1, 4)]
    sm[1] = [softmax(r, 3) for r in range(1, 4)]
    population = [[0, 1, 2], [0, 1, 2]]
    for i in range(3):
        cointoss = random.randint(0, 1)
        index = random.choices(population[cointoss], sm[cointoss])[0]
        sm[cointoss][index] = 0
        result += [(new_pair[cointoss][index], coin_to_ranker(cointoss))]
        for j in range(3):
            if (new_pair[cointoss][index] == new_pair[1 - cointoss][j]):
                sm[1 - cointoss][j] = 0       
    return result

def decide_winner(interl, clicks):
    """
    Takes an interleaved list and a click pattern and returns the winner: E, P or NW (no winner)
    """
    clicks_E = 0
    clicks_P = 0
    for i in range(3):
        if (clicks[i] == 1):
            if (interl[i][1] == "E"):
                clicks_E += 1
            elif (interl[i][1] == "P"):
                clicks_P += 1
    if (clicks_E > clicks_P):
        return "E"
    elif (clicks_E < clicks_P):
        return "P"
    else:
        return "NW"
    
def decide_winner_marg(interl, ranking_pair, clicks):
    """
    Returns winner of probabilistic interleaving according to the marginalised estimate
    Note that the probability of an assignment given an interleaved list is proportional
    to the probability of a list given an assignment, and thus only the latter is computed
    """
    prob_win = [0,0] # unnormalised 'probability' that system E or P wins
    
    # loop over assignments
    for assignment in [[a,b,c] 
                       for a in [0,1] 
                       for b in [0,1] 
                       for c in [0,1]]:
        prob_list = 1 # probability of interleaved list given assignment
        click_count = [0,0] # count how often each system's docs are clicked
        sm = [[softmax(r, 3) for r in range(1, 4)], 
              [softmax(r, 3) for r in range(1, 4)]]
        # loop over list positions
        for i in range(3):
            doc = interl[i][0]
            ass = assignment[i]
            cli = clicks[i]
            # compute probability of list given assignment
            if doc not in ranking_pair[ass]:
                # assignment is not possible
                prob_list = 0
                break
            else:
                # count clicks
                click_count[ass] += cli
                # update probability
                index = ranking_pair[ass].index(doc)
                prob_list *= sm[ass][index]
                # normalise softmax
                sm[ass][index] = 0
                sum_sm = sum(sm[ass])
                if sum_sm != 0:
                    sm[ass] = [s/sum_sm for s in sm[ass]]
                # update softmax of other list if necessary
                if doc in ranking_pair[1-ass]:
                    index = ranking_pair[1-ass].index(doc)
                    sm[1-ass][index] = 0
                    sum_sm = sum(sm[1-ass])
                    if sum_sm != 0:
                        sm[1-ass] = [s/sum_sm for s in sm[1-ass]]
        # assign win probability to appropriate system
        if click_count[0] > click_count[1]:
            prob_win[0] += prob_list
        elif click_count[1] > click_count[0]:
            prob_win[1] += prob_list
            
    # return winner
    if (prob_win[1] > prob_win[0]):
        return "P"
    elif (prob_win[1] < prob_win[0]):
        return "E"
    else:
        return "NW"
```

## Step 4: Simulate User Clicks

```python
def yandex_log_parser():
    """
    Parses the Yandex Click Log File and returns a list with for each session a list with a 1
    a list of the form [rank1_clicked, rank2_clicked, ..., rank10_clicked] where ranki_clicked
    is either 1 if clicked or 0 if not clicked.
    """
    sessionID = 0
    links = [-1 for i in range(10)] #dummy list
    clicks = [0 for i in range(10)]
    list_clicks = []
    with open("YandexRelPredChallenge.txt") as f:
        for line in f:
            words = line.split()
            sessionID_old = sessionID
            sessionID = int(words[0])
            if sessionID>sessionID_old:
                list_clicks.append(clicks)
                clicks = [0 for i in range(10)]
            recordType = words[2]
            if recordType=="Q":
                links = [int(l) for l in words[5:]]
            elif recordType=="C":
                link_clicked = int(words[3])
                if link_clicked in links:
                    rank_clicked = links.index(link_clicked)
                    clicks[rank_clicked] =1
    return(list_clicks)

def em():
    """
    Expectation-maximization method for determining the parameters alpha and gamma, 
    using training data from the yandex_log_parser().
    Using the update rules from [1].
    """
    diffs = [1]*11 # differences in parameters (first alpha, then ten gammas)
    diff_term = 1e-10 # minimal difference for termination
    max_iter = 100
    
    gamma = [0.5 for x in range(10)] # initial probability for each rank whether observed
    alpha = 0.2 # initial probability clicked if not relevant
    epsilon = 0.05
    tolerance = 0.01
    click_log = yandex_log_parser()
    i = 0
    while i < max_iter and max(diffs) > diff_term:
        i+=1
        alpha_old = alpha
        gamma_old = gamma.copy()
        total = [2 for x in range(10)]
        for j in click_log:
            for rank in range(len(j)):
                gamma_value = gamma[rank]/total[rank]
                alpha_value = alpha/sum(total)
                if j[rank]==1:
                    gamma[rank] += 1
                    alpha += 1
                else:
                    alpha += (1-gamma_value)*alpha_value/(1-gamma_value*alpha_value)
                    gamma[rank] += (1-alpha_value)*gamma_value/(1-gamma_value*alpha_value)
                total[rank] += 1
        # New alpha and gamma
        alpha = alpha/sum(total)
        diffs[0] = abs(alpha-alpha_old)
        for x in range(len(gamma)):
            gamma[x] = gamma[x]/total[x]
            diffs[x+1] = abs(gamma[x]-gamma_old[x])
    return (alpha,gamma)

def click_probabilities(l, alpha, gamma):
    """
    Returns the probability of each position being clicked based on ranked relevance list l,
    and parameters alpha and gamma
    """
    return [abs(alpha - (1 - l[i]))*gamma[i] for i in range(3)]

def produce_clicks(list, alpha, gamma):
    """
    Performs the Position-Based Click Model (PBM)
    Returns which positions are clicked based on the ranked relevance list and parameters alpha and gamma
    """
    probabilities = click_probabilities(list, alpha, gamma)    
    return [np.random.binomial(1, probabilities[i]) for i in range(3)]

def produce_clicks_random(list, theta):
    """
    Performs the Random Click Model
    Returns which positions are clicked based on parameter theta (probability to click)
    """
    return [np.random.binomial(1, theta) for i in range(3)]
```

## Step 5: Simulate Interleaving Experiment

```python
## Simulation of Interleaving Experiment


def estimate_win_proportion(ranking_pair, interleave_function, click_function, alpha, gamma, theta, k):
    """
    Runs experiment to estimate the win proportion of E on a single pair
    Interleaving and simulating clicks is done k times
    The ranking pair, functions to interleave and click and their parameters are arguments to this function
    Note that this function does not determine the winner by the marginalised estimation of probabilistic interleaving
    """
    wins_E = 0
    wins_P = 0
    for i in range(k):
        interl = interleave_function(ranking_pair)
        relevance_list = get_relevance_list([elem[0] for elem in interl])
        if (click_function == produce_clicks):
            clicks = click_function(relevance_list, alpha, gamma)
        elif (click_function == produce_clicks_random):
            clicks = click_function(relevance_list, theta)
        winner = decide_winner(interl, clicks)
        if (winner == "E"):
            wins_E += 1
        elif (winner == "P"):
            wins_P += 1
            
    if (wins_E + wins_P > 0):
        return wins_E / (wins_E + wins_P)
    else:
        return 0.5 # if all simulations resulted in a tie, return 0.5
    
    
# Determines win proportion of experimental system based on the marginalised estimator of probabilistic interleaving
def estimate_win_proportion_marginalised(ranking_pair, interleave_function, click_function, alpha, gamma, theta, k):
    """
    Runs experiment to estimate the win proportion of E on a single pair based on the marginalised estimation
    of probabilistic interleaving
    Interleaving and simulating clicks is done k times
    The ranking pair, functions to interleave and click and their parameters are arguments to this function
    """
    wins_E = 0
    wins_P = 0
    for i in range(k):
        # interleave 
        interl = interleave_function(ranking_pair)
        # simulate clicks
        relevance_list = get_relevance_list([elem[0] for elem in interl])
        if (click_function == produce_clicks):
            clicks = produce_clicks(relevance_list, alpha, gamma)
        elif (click_function == produce_clicks_random):
            clicks = produce_clicks_random(relevance_list, theta)
        # decide winning system
        winner = decide_winner_marg(interl, ranking_pair, clicks)
        if (winner == "E"):
            wins_E += 1
        elif (winner == "P"):
            wins_P += 1
    
    if (wins_E + wins_P > 0):
        return wins_E / (wins_E + wins_P)
    else:
        return 0.5 # if all simulations resulted in a tie, return 0.5
    
def run_interleaving_experiment(deltaERRs, estimator, interleaver, click_function, alpha, gamma, theta, k):
    """
    Runs a full interleaving experiment on ranking pairs orginised in bins (deltaERRs), using an estimator
    that runs the experiment for each pair based on an interleaver and click function
    Returns for each bin a tuple containing the minimum, maximum and median number of the estimated minimum 
    number of impressions N, and the percentage of pairs that resulted in a win probability of E lower than 0.5
    Note that in this case, N would be estimated as infinity
    """
    result = []
    counter = 0
    total_pairs = sum([len(d) for d in deltaERRs])
    print_at = np.ceil(np.linspace(total_pairs/5, total_pairs, 5))
    start_time = time.time()
        
    for i in range(len(deltaERRs)): # loop over bins
        if (len(deltaERRs[i])> 0):
            Ns = []
            count_Ninf = 0 # counts how often N is infinite (p1<=0.5)
            for pair in deltaERRs[i]: # loop over pairs in bin
                p1 = estimator(pair, interleaver, click_function, alpha, gamma, theta, k)
                counter += 1
                if counter in print_at:
                    print("\t{:.0f}%\t({:.0f} secs to go)".format(
                        counter/total_pairs*100,
                        math.ceil((time.time()-start_time)/counter*total_pairs-time.time()+start_time)))
                N = compute_sample_size(p1)
                Ns += [N]
                if (N == np.inf):
                    count_Ninf += 1
                    
            minimum = min(Ns)
            maximum = max(Ns)
            median = statistics.median(Ns)
            result += [(minimum, median, maximum, count_Ninf/len(deltaERRs[i]))]
        else:
            result += [(None,None,None, None)]
    return result
```

## Step 6: Compute Sample Size

```python
def compute_sample_size(p1):
    """
    Returns the required minimal sample size according to the power analysis of the 
    one-sided one-sample proportion test with alpha 0.05, beta 0.1 and p0 0.5
    """
    if p1<=0.5:
        return np.inf
    a = 0.05
    b = 0.1
    p0 = 0.5
    delta = abs(p1-p0)
    N_intermediate = ((norm.ppf(1 - a)*math.sqrt(p0*(1-p0)) + 
                      norm.ppf(1 - b)*math.sqrt(p1*(1 - p1)))
                      /delta)**2
    N = N_intermediate + 1/delta
    return math.ceil(N)
```

## Tables and Plots

```python
def results_table(results):
    """
    Shows a table of min, median and max number of impressions in each bin, and the
    percentage of experiments per bin that led to infinite N
    """
    xlabels = ["min", "median", "max", "infinite"]
    ylabels = []
    table = []
    i = 0
    for row in results:
        ylabels.append("{:.2f} - {:.2f}".format(.05 if i==0 else i/10, (i+1)/10))
        i += 1
        if row[0] is None:
            table.append(["-"]*4)
        else:
            table.append(["{:.0f}".format(row[0]),
                          "{:.1f}".format(row[1]),
                          "{:.0f}".format(row[2]),
                          "{:.1f}%".format(row[3]*100)])
            
    return pandas.DataFrame(table, ylabels, xlabels)


def plot_medians(td, pr, pr_mar):
    """
    Displays plot of median number of impressions per interleaving method
    """
    data_dict = {
        "team draft": td,
        "probab": pr,
        "probab, marg": pr_mar}
    
    for name in data_dict.keys():
        medians = [row[1] for row in data_dict[name]]
        plt.plot([x+1 for x in range(10)],medians, "o-", label=name)
    
    plt.xlim((1,10))
    plt.legend()
    plt.ylabel("median")
    plt.xlabel("bin")
    plt.title("Median impressions per bin")
    plt.show()

def plot_pairs_histogram(pairs_in_bins):
    """
    Displays a histogram of the number of pairs in each bin
    """
    height = [len(pair) for pair in pairs_in_bins]
    bins = [x+1 for x in range(10)]
    print(len(height), len(bins))
    plt.bar(bins, height)
    plt.xlabel("bin")
    plt.ylabel("number of pairs")
    plt.title("Number of ranking pairs in each bin")
    plt.show()
```

```python
random.seed(42)
np.random.seed(42)
print("Creating all ranking pairs")
pairs = create_ranking_pairs() #list of all distinct ranking pairs
print("\t{} pairs created\n".format(len(pairs)))

print("Computing deltaERR for each pair and assigning pairs to bins")
deltaERRs = divide_pairs_over_bins(pairs) #list of all pairs (tuples of two lists) divided over the 10 bins
print("\t{} non-empty bins created with in total {} pairs\n".format(
    len([bin for bin in deltaERRs if len(bin) > 0]),
    sum([len(p) for p in deltaERRs])))

print("Performing EM to estimate position based click model parameters")
(alpha, gamma) = em()
epsilon = 0.05
alpha_em = alpha
alpha = 1 - epsilon    # alpha is manually determined
gamma = gamma[:3]      # only first 3 gammas are used
print("\talpha: {:.3f} (manually set after EM, result from EM: {:.3f})".format(alpha, alpha_em))
print("\tgammas: {}\n".format(", ".join(["{:.3f}".format(g) for g in gamma])))

print("Setting random click model parameter")
theta = 0.5
print("\ttheta: {:.3f}\n".format(theta))

k = 10000
print("Starting online experiments with k={}".format(k))

# Experiment 1: team draft interleaving, position-based click model
estimator = estimate_win_proportion
interleaver = td_interleave
clicker = produce_clicks
print("\nRunning experiment 1: team draft interleaving & position-based click model")
results_td_pb = run_interleaving_experiment(deltaERRs, 
                                    estimator, 
                                    interleaver, 
                                    clicker, 
                                    alpha, gamma, theta, k)

# Experiment 2: team draft interleaving, random click model
estimator = estimate_win_proportion
interleaver = td_interleave
clicker = produce_clicks_random
print("\nRunning experiment 2: team draft interleaving & random click model")
results_td_rn = run_interleaving_experiment(deltaERRs, 
                                    estimator, 
                                    interleaver, 
                                    clicker, 
                                    alpha, gamma, theta, k)

# Experiment 3: probabilistic interleaving, position-based click model
estimator = estimate_win_proportion
interleaver = pr_interleave
clicker = produce_clicks
print("\nRunning experiment 3: probabilistic interleaving & position-based click model")
results_pr_pb = run_interleaving_experiment(deltaERRs, 
                                    estimator, 
                                    interleaver, 
                                    clicker, 
                                    alpha, gamma, theta, k)

# Experiment 4: probabilistic interleaving, random click model
estimator = estimate_win_proportion
interleaver = pr_interleave
clicker = produce_clicks_random
print("\nRunning experiment 4: probabilistic interleaving & random click model")
results_pr_rn = run_interleaving_experiment(deltaERRs, 
                                    estimator, 
                                    interleaver, 
                                    clicker, 
                                    alpha, gamma, theta, k)

# Experiment 5: marginalised probabilistic interleaving, position-based click model
estimator = estimate_win_proportion_marginalised
interleaver = pr_interleave
clicker = produce_clicks
print("\nRunning experiment 5: marginalised probabilistic interleaving & position-based click model")
results_pr_pb_mar = run_interleaving_experiment(deltaERRs, 
                                    estimator, 
                                    interleaver, 
                                    clicker, 
                                    alpha, gamma, theta, k)

# Experiment 6: marginalised probabilistic interleaving, random click model
estimator = estimate_win_proportion_marginalised
interleaver = pr_interleave
clicker = produce_clicks_random
print("\nRunning experiment 6: marginalised probabilistic interleaving & random click model")
results_pr_rn_mar = run_interleaving_experiment(deltaERRs, 
                                    estimator, 
                                    interleaver, 
                                    clicker, 
                                    alpha, gamma, theta, k)

print("Showing results\n\n")

print("Table 1: team draft interleaving & position-based click model")
print(results_table(results_td_pb))

print("\nTable 2: team draft interleaving & random click model")
print(results_table(results_td_rn))
    
print("\nTable 3: probabilistic interleaving & position-based click model")
print(results_table(results_pr_pb))

print("\nTable 4: probabilistic interleaving & random click model")
print(results_table(results_pr_rn)) 

print("\nTable 5: marginalised probabilistic interleaving & position-based click model")
print(results_table(results_pr_pb_mar))

print("\nTable 6: marginalised probabilistic interleaving & random click model")
print(results_table(results_pr_rn_mar))

print("\nFigure 1: Median impressions per bin, position-based click model")
plot_medians(results_td_pb, results_pr_pb, results_pr_pb_mar)
      
print("\nFigure 2: Median impressions per bin, random click model")
plot_medians(results_td_rn, results_pr_rn, results_pr_rn_mar)

print("\nFigure 3: Number of pairs per bin")
plot_pairs_histogram(deltaERRs)
```

## Analysis and Conclusion
The main conclusion is that the offline evaluation can inform the design of the online comparison of two ranking algorithms in a meaningful way, given the assumptions made in the experiment. The strongest assumptions are that PBM accurately describes user click behaviour; that the same set of queries are used in the offline and online evaluation; and that there is no query variance. Figure 1 shows a correlation between the result of the offline evaluation ($\Delta$ERR) and the median number of impressions needed to significantly compare two ranking algorithms. As is expected, an experiment with an algorithm pair with a larger $\Delta$ERR requires less impressions to make a significant conclusion about the relative performance.

### Click models
Experiments using the random click model are used as sanity check. About half of the pairs in each bin result in an infinite N. This confirms the check, because systems E and P have equal chances of winning. 

With the results for the position-based click model, there are significant differences between the interleaving methods, as can be seen in Table 1, 3 and 5.
Probabilistic marginalised interleaving seems to require less impressions to show a significant difference in the lower bins 1 and 2 (see Figure 1). With higher $\Delta$ERR values the difference is small.

We first analyze the results of the random click-model. In this model, there is a fixed probability of, in our case, 50% that a document is clicked. This probability is independent from the rank and the relevance of a document. It therefore makes sense that neither the $\Delta$ERR, nor the interleaving method has any significant influence on the results. This explains why we consistently see approximately the same numbers in the tables.

With the position-based click model, the click-probability of a document depends on both its rank and its relevance. In particular, the dependence on the relevance effectively means that a ranker will receive more clicks if it more often produces relevant results. If those relevant results are higher in the ranking, they receive even more clicks. In our case, the experimental algorithm E more often produces relevant, high-ranked results in the cases where the $\Delta$ERR is big, and that means that differences in the algorithms are more easily detectable in a statistically significant way. Therefore, we could expect a priori that the results for N drop with increasing Delta-ERR. That’s also what we observe in the tables.

### Team draft interleaving
About team-draft interleaving: This follows the trend already described in the click-models section. The required number of impressions drops almost exponentially with an increasing ΔERR. Furthermore, we observe that in the first bin, roughly 4.4% of pairs lead to a value of infinity. This can potentially be explained by the insufficient sensitivity of team draft interleaving that moves the average win-proportion of E closer to 0.5.

### Probabilistic interleaving
Based on Figure 1, team draft interleaving seems to perform slightly better than probabilistic interleaving. This is against our expectations. Probabilistic interleaving was developed to be more sensitive which means that is it better at distinguishing between systems that are very close (and are not distinguished by team draft)[3]. The source of the differences could be further investigated. 

Probabilistic marginalised interleaving seems to perform better than probabilistic interleaving. This is what we expected. In probabilistic interleaving, luck plays a larger role in winning a simulation. Marginalised interleaving looks at the interleaved list and the click pattern, and then decides which system most probably contributed the clicked documents. Normal probabilistic interleaving decides based on the accidental assignment per simulation. Thus, by marginalising we increase the probability of the better system winning, and therefore need a smaller amount of impressions. This effect is larger for systems with smaller differences, as is confirmed by Figure 1. If the differences are already large, the worse system can already hardly win and the number of required impressions is generally small.

Nevertheless, we find the following difference unexpected: Probabilistic marginalised interleaving produced more max-values of infinity than usual probabilistic interleaving: 6.7% and 4.5% in the first two bins, compared to no cases at all. While it can be that $\Delta$ERR favors one ranker that is actually not better on the online-evaluation, the difference between the two versions of probabilistic interleaving requires an explanation. Further experiments are needed to explain this phenomenon.

## Possible improvements

### Different click-models

A different click model could be implemented in order to potentially better simulate actual user behaviour. Examples of widely used click models are the Dependent Click Model (DCM), Click Chain Model (CCM), and a Dynamic Bayesian Network Model (DBN)[1].

### Better parameter-estimation

To find the parameters of the click model, EM is used to infer the examination (gamma) and the attractiveness (alpha) probabilities. Currently, we use a fixed constant epsilon to define how probable it is that a user clicks on a non-relevant document when he sees it, because it is difficult to infer the value for alpha for each possible document-query pair and we have no information about which documents are relevant and which are not. It could be tried to acquire click logs where multiple sessions occur with the same document-query pairs to compute meaningful alpha values for these document-query pairs. The alpha inferred for documents which are clicked a significant number of times can inform a sensible choice of alpha for relevant documents. This could be done by for instance calculating the average over all these document-query pairs with enough clicks. These click logs could be acquired by setting up an experiment where users have to choose the best documents given a query.

### Closed-form solutions

A closed form solution can be found, which gives a more accurate and computationally efficient method for finding the number of impressions. This has also been suggested in [2] and our own description of the experiment (see the appendix of part A of the programming assignment). With a closed-form solution the variance is reduced and we therefore get more precise values about the number of impressions to choose. This can only be approached in the current experiment by choosing a very high value for *k*, which has computational disadvantages compared to the closed-form solution.

### Beyond the no-query-variability assumption

Currently, the assumption is made that there is no query variability, which means that all queries would lead to the same document ranking. It is also assumed that the queries in the offline and online experiment are the same. The assumptions make the experiment design easier, but at the expense of less realistically modelling real search algorithms. Finding a way to take the query variability into account, as is also suggested by [2], would be an improvement for the experimental design.
